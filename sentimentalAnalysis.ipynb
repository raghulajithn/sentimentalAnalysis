PREPROCESSING

```python
df.rename(columns={'review':'text'}, inplace = True)
df
```

```python
def cleaning(text):
    # converting to lowercase, removing URL links, special characters, punctuations...
    text = text.lower() # converting to lowercase
    text = re.sub('https?://\S+|www\.\S+', '', text) # removing URL links
    text = re.sub(r"\b\d+\b", "", text) # removing number
    text = re.sub('<.*?>+', '', text) # removing special characters,
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text) # punctuations
    text = re.sub('\n', '', text)
    text = re.sub('[’“”…]', '', text)

    #removing emoji:
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    text = emoji_pattern.sub(r'', text)

   # removing short form:

    text=re.sub("isn't",'is not',text)
    text=re.sub("he's",'he is',text)
    text=re.sub("wasn't",'was not',text)
    text=re.sub("there's",'there is',text)
    text=re.sub("couldn't",'could not',text)
    text=re.sub("won't",'will not',text)
    text=re.sub("they're",'they are',text)
    text=re.sub("she's",'she is',text)
    text=re.sub("There's",'there is',text)
    text=re.sub("wouldn't",'would not',text)
    text=re.sub("haven't",'have not',text)
    text=re.sub("That's",'That is',text)
    text=re.sub("you've",'you have',text)
    text=re.sub("He's",'He is',text)
    text=re.sub("what's",'what is',text)
    text=re.sub("weren't",'were not',text)
    text=re.sub("we're",'we are',text)
    text=re.sub("hasn't",'has not',text)
    text=re.sub("you'd",'you would',text)
    text=re.sub("shouldn't",'should not',text)
    text=re.sub("let's",'let us',text)
    text=re.sub("they've",'they have',text)
    text=re.sub("You'll",'You will',text)
    text=re.sub("i'm",'i am',text)
    text=re.sub("we've",'we have',text)
    text=re.sub("it's",'it is',text)
    text=re.sub("don't",'do not',text)
    text=re.sub("that´s",'that is',text)
    text=re.sub("I´m",'I am',text)
    text=re.sub("it’s",'it is',text)
    text=re.sub("she´s",'she is',text)
    text=re.sub("he’s'",'he is',text)
    text=re.sub('I’m','I am',text)
    text=re.sub('I’d','I did',text)
    text=re.sub("he’s'",'he is',text)
    text=re.sub('there’s','there is',text)


    return text

dt = df['text'].apply(cleaning)
```

```python
dt = pd.DataFrame(dt)
dt['sentiment']=df['sentiment']
dt
```

```python
# remove stop word:
stop_words = set(stop_words)  # Convert stop_words list to a set for faster lookup

dt['no_sw'] = dt['text'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
```

```python
dt
```

```python
#Working with the most Frequent Words:
from collections import Counter
cnt = Counter()
for text in dt["no_sw"].values:
    for word in text.split():
        cnt[word] += 1
cnt.most_common(10)
temp = pd.DataFrame(cnt.most_common(10))
temp.columns=['word', 'count']
temp
```

```python
px.bar(temp, x="count", y="word", title='Commmon Words in Text', orientation='h',
             width=700, height=700)
```

```python
# Remove the most frequent words:
FREQWORDS = set([w for (w, wc) in cnt.most_common(10)])
def remove_freqwords(text):
    """custom function to remove the frequent words"""
    return " ".join([word for word in str(text).split() if word not in FREQWORDS])
dt["wo_stopfreq"] = dt["no_sw"].apply(lambda text: remove_freqwords(text))
dt.head()
```

```python
# Lemmatization: Lemmatization is converting the word to its base form or lemma by removing affixes from the inflected words.
# It helps to create better features for machine learning and NLP models hence it is an important preprocessing step.

nltk.download('wordnet')
wordnet_lem = WordNetLemmatizer()

dt['wo_stopfreq_lem'] = dt['wo_stopfreq'].apply(wordnet_lem.lemmatize)
dt
```

```python
# create the cleaned data for the train-test split:
nb=dt.drop(columns=['text','no_sw', 'wo_stopfreq'])
nb.columns=['sentiment','review']
nb.sentiment = [0 if each == "negative" else 1 for each in nb.sentiment]
nb
```

```python
tokenized_review=nb['review'].apply(lambda x: x.split())
tokenized_review.head(5)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from nltk.tokenize import RegexpTokenizer

# Initialize the RegexpTokenizer
tokenizer = RegexpTokenizer(r'[a-zA-Z0-9]+')

# Initialize the CountVectorizer with stop_words='english'
cv = CountVectorizer(stop_words='english', ngram_range=(1, 1), tokenizer=tokenizer.tokenize)

# Fit and transform the text data
text_counts = cv.fit_transform(nb['review'])

# Remove the word 'not' from the vocabulary
vocabulary = cv.vocabulary_
vocabulary.pop('not', None)
cv.set_params(vocabulary=vocabulary)
text_counts = cv.transform(nb['review'])
```

MODEL IMPLEMENTATION

```python
from sklearn.model_selection import train_test_split
X=text_counts
y=nb['sentiment']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20,random_state=30)
```

```python
from sklearn.naive_bayes import ComplementNB
from sklearn.metrics import classification_report, confusion_matrix
CNB = ComplementNB()
CNB.fit(X_train, y_train)

from sklearn import metrics
predicted = CNB.predict(X_test)
accuracy_score = metrics.accuracy_score(predicted, y_test)

print('ComplementNB model accuracy is',str('{:04.2f}'.format(accuracy_score*100))+'%')
print('------------------------------------------------')
print('Confusion Matrix:')
print(pd.DataFrame(confusion_matrix(y_test, predicted)))
print('------------------------------------------------')
print('Classification Report:')
print(classification_report(y_test, predicted))
```

```python
from sklearn.naive_bayes import MultinomialNB

MNB = MultinomialNB()
MNB.fit(X_train, y_train)

predicted = MNB.predict(X_test)
accuracy_score = metrics.accuracy_score(predicted, y_test)

print('MultinominalNB model accuracy is',str('{:04.2f}'.format(accuracy_score*100))+'%')
print('------------------------------------------------')
print('Confusion Matrix:')
print(pd.DataFrame(confusion_matrix(y_test, predicted)))
print('------------------------------------------------')
print('Classification Report:')
print(classification_report(y_test, predicted))
```

```python
from sklearn.naive_bayes import BernoulliNB

BNB = BernoulliNB()
BNB.fit(X_train, y_train)

predicted = BNB.predict(X_test)
accuracy_score_bnb = metrics.accuracy_score(predicted,y_test)

print('BernoulliNB model accuracy = ' + str('{:4.2f}'.format(accuracy_score_bnb*100))+'%')
print('------------------------------------------------')
print('Confusion Matrix:')
print(pd.DataFrame(confusion_matrix(y_test, predicted)))
print('------------------------------------------------')
print('Classification Report:')
print(classification_report(y_test, predicted))
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML, Image

def handle_button_click(button):
    if user_input_widget.value.lower() == 'exit':
        return
    elif user_input_widget.value != '':
        new_text = [str(user_input_widget.value)]  # Convert to string

        # Custom logic for sentiment prediction
        text = user_input_widget.value.lower()
        if 'good' in text and ('not' not in text or text.index('good') < text.index('not')):
            predicted_sentiment = 'positive'
        elif 'bad' in text and ('not' not in text or text.index('bad') < text.index('not')):
            predicted_sentiment = 'negative'
        else:
            new_text_features = tfidf.transform(new_text)
            predicted_sentiment = CNB.predict(new_text_features)
            predicted_sentiment = 'negative' if predicted_sentiment[0] == 0 else 'positive'

        # Set the color based on the predicted sentiment
        color = 'green' if predicted_sentiment == 'positive' else 'red'

        # Set the smiley image based on the predicted sentiment
        image_path = '/content/drive/MyDrive/happy smiley.png' if predicted_sentiment == 'positive' else '/content/drive/MyDrive/sad smiley.png'
        smiley_image = Image(filename=image_path, width=250, height=250)  # Adjust the width and height values as needed

        # Create the HTML element with the sentiment value in the specified color and smiley image
        sentiment_html = f"<span style='color:{color}; font-size:20px;'>{predicted_sentiment}</span><br>"

        # Display the predicted sentiment and smiley image in the output area
        with output_area:
            clear_output()
            display(HTML(f"<h2>Predicted sentiment: {sentiment_html}</h2>"))
            display(smiley_image)

        user_input_widget.value = ''  # Clear the input text

button_widget = widgets.Button(description='Predict', layout=widgets.Layout(width='125px', height='30px',  border='2px solid white'))
button_widget.style.button_color = 'purple'  # Set the button color

user_input_widget = widgets.Textarea(
    value='',
    placeholder='Enter a text to predict its sentiment (or type "exit" to quit)',
    description='Text:',
    layout=widgets.Layout(width='50%', height='30px')  # Set the width and height of the text box
)

output_area = widgets.Output()
display(widgets.HBox([user_input_widget, button_widget]))  # Display the input and button in a horizontal box
display(output_area)

button_widget.on_click(handle_button_click)
```
